```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score

# -------------------------------
# Settings
# -------------------------------
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
N_SAMPLES = 300
TRUE_CENTERS = 3        # used only for data generation
CLUSTER_STD = 1.2

# -------------------------------
# 1) Generate well-separated dataset
# -------------------------------
X, y_true = make_blobs(
    n_samples=N_SAMPLES,
    centers=TRUE_CENTERS,
    cluster_std=CLUSTER_STD,
    random_state=RANDOM_STATE
)

# Interpret features as "Annual Income" and "Spending Score"
# (this is just axis labels; the blob generator returns two features)
# -------------------------------
# 2) K-Means (NumPy from scratch)
# -------------------------------
def kmeans(X, k, max_iters=300, tol=1e-4, random_state=None):
    if random_state is not None:
        rng = np.random.RandomState(random_state)
    else:
        rng = np.random

    # initialize centroids by sampling k unique points
    centroids = X[rng.choice(len(X), k, replace=False)].astype(float)

    for it in range(max_iters):
        # distances: (n_samples, k)
        distances = np.linalg.norm(X[:, None, :] - centroids[None, :, :], axis=2)
        labels = np.argmin(distances, axis=1)

        new_centroids = np.zeros_like(centroids)
        for i in range(k):
            members = X[labels == i]
            if len(members) == 0:
                # handle empty cluster: reinitialize centroid to a random data point
                new_centroids[i] = X[rng.randint(0, len(X))]
            else:
                new_centroids[i] = members.mean(axis=0)

        # convergence check
        if np.allclose(centroids, new_centroids, atol=tol):
            centroids = new_centroids
            break

        centroids = new_centroids

    # final labels and WCSS
    distances = np.linalg.norm(X[:, None, :] - centroids[None, :, :], axis=2)
    labels = np.argmin(distances, axis=1)
    wcss = np.sum((X - centroids[labels]) ** 2)

    return centroids, labels, wcss

# -------------------------------
# 3) Run Elbow (WCSS) and Silhouette (K=2..10)
# -------------------------------
K_range = range(2, 11)
wcss_list = []
silhouette_list = []

for k in K_range:
    centroids, labels, wcss = kmeans(X, k, random_state=RANDOM_STATE)
    wcss_list.append(wcss)

    # silhouette_score exists only when there are >= 2 clusters and < n_samples clusters
    try:
        sil = silhouette_score(X, labels)
    except Exception:
        sil = float('nan')
    silhouette_list.append(sil)

# Plot Elbow and Silhouette side-by-side
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
ax[0].plot(list(K_range), wcss_list, marker='o')
ax[0].set_xlabel("Number of Clusters (K)")
ax[0].set_ylabel("WCSS")
ax[0].set_title("Elbow Method (WCSS)")

ax[1].plot(list(K_range), silhouette_list, marker='o')
ax[1].set_xlabel("Number of Clusters (K)")
ax[1].set_ylabel("Silhouette Score")
ax[1].set_title("Silhouette Analysis")
plt.tight_layout()
plt.show()

# -------------------------------
# 4) Choose best K and fit final model
#    (choose K with highest silhouette; tie-breaker: lower K)
# -------------------------------
best_idx = int(np.nanargmax(silhouette_list))
best_k = list(K_range)[best_idx]
print(f"Chosen best K by silhouette score: K = {best_k}")

centroids, labels, final_wcss = kmeans(X, best_k, random_state=RANDOM_STATE)

# -------------------------------
# 5) Print centroids and first 10 labels
# -------------------------------
print("\nFinal Cluster Centroids (format: [Income, Spending]):")
for i, c in enumerate(centroids):
    print(f"Cluster {i}: {c}")

print("\nFirst 10 data points and assigned cluster:")
for i in range(10):
    print(f"Point {i}: {X[i]} -> Cluster {labels[i]}")

# -------------------------------
# 6) Cluster characteristics & interpretation
# -------------------------------
cluster_info = []
for i in range(best_k):
    members = X[labels == i]
    size = len(members)
    mean_income = members[:, 0].mean()
    mean_spending = members[:, 1].mean()
    std_income = members[:, 0].std()
    std_spending = members[:, 1].std()
    cluster_info.append({
        'cluster': i,
        'size': size,
        'mean_income': mean_income,
        'mean_spending': mean_spending,
        'std_income': std_income,
        'std_spending': std_spending
    })

# derive thresholds (33rd and 66th percentiles) to describe low/medium/high
income_q1, income_q2 = np.percentile(X[:, 0], [33, 66])
spend_q1, spend_q2 = np.percentile(X[:, 1], [33, 66])

def describe_value(val, q1, q2):
    if val <= q1:
        return "low"
    elif val <= q2:
        return "medium"
    else:
        return "high"

print("\nCluster summary and simple business interpretation:")
for info in cluster_info:
    inc_desc = describe_value(info['mean_income'], income_q1, income_q2)
    sp_desc = describe_value(info['mean_spending'], spend_q1, spend_q2)
    print(
        f"Cluster {info['cluster']}: size={info['size']}, "
        f"mean_income={info['mean_income']:.1f} ({inc_desc}), "
        f"mean_spending={info['mean_spending']:.1f} ({sp_desc})"
    )
    # Example interpretation sentence
    print(f"  -> Interpretation: This cluster is {inc_desc}-income and {sp_desc}-spending customers.\n")

# -------------------------------
# 7) Final Visualization
# -------------------------------
plt.figure(figsize=(8,6))
scatter = plt.scatter(X[:, 0], X[:, 1], c=labels, s=40)
plt.scatter(centroids[:, 0], centroids[:, 1], s=220, marker='X', edgecolor='k', linewidth=1.5)
plt.xlabel("Annual Income (feature 0)")
plt.ylabel("Spending Score (feature 1)")
plt.title(f"Final K-Means Clustering (K={best_k})")
plt.grid(alpha=0.2)

# annotate centroids
for i, c in enumerate(centroids):
    plt.text(c[0], c[1], f"  C{i}", fontsize=12, weight='bold')

plt.show()

# -------------------------------
# 8) Save final results (optional)
# -------------------------------
# To save labels + centroids to CSV uncomment the lines below:
# import pandas as pd
# df_out = pd.DataFrame(X, columns=['income', 'spending'])
# df_out['cluster'] = labels
# df_out.to_csv('kmeans_results.csv', index=False)
# print("Saved clustering result to kmeans_results.csv")
```
