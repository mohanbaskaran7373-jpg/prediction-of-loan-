```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# 1. Generate Realistic Dataset
# -------------------------------
np.random.seed(42)

age = np.random.randint(18, 60, 200)
income = np.random.randint(15000, 100000, 200)
spending = np.random.randint(1, 100, 200)

X = np.column_stack((income, spending))

# -------------------------------
# 2. K-Means from Scratch
# -------------------------------
def kmeans(X, k, iterations=100):
    centroids = X[np.random.choice(len(X), k, replace=False)]

    for _ in range(iterations):
        distances = np.linalg.norm(X[:, None] - centroids, axis=2)
        labels = np.argmin(distances, axis=1)

        new_centroids = np.array([
            X[labels == i].mean(axis=0) for i in range(k)
        ])

        if np.allclose(centroids, new_centroids):
            break

        centroids = new_centroids

    return centroids, labels

# -------------------------------
# 3. Elbow Method
# -------------------------------
wcss = []
for k in range(1, 7):
    centroids, labels = kmeans(X, k)
    cost = np.sum((X - centroids[labels]) ** 2)
    wcss.append(cost)

plt.plot(range(1,7), wcss, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()

# -------------------------------
# 4. Final Model (K = 3)
# -------------------------------
centroids, labels = kmeans(X, 3)

# Print centroids
print("Final Cluster Centroids:")
print(centroids)

# First 10 data point assignments
print("\nFirst 10 data point cluster labels:")
for i in range(10):
    print(X[i], "-> Cluster", labels[i])

# -------------------------------
# 5. Visualization
# -------------------------------
plt.scatter(X[:,0], X[:,1], c=labels)
plt.scatter(centroids[:,0], centroids[:,1], s=200)
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Customer Segmentation using K-Means")
plt.show()
```
