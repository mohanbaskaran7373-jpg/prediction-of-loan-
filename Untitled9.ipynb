```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler


np.random.seed(42)
n_samples = 500
data = np.concatenate([
    np.random.normal(loc=[10, 5], scale=1.5, size=(20, 2)),   # Cluster 1: low cost, low dist
    np.random.normal(loc=[20, 15], scale=2, size=(20, 2)),    # Cluster 2
    np.random.normal(loc=[30, 25], scale=1.8, size=(20, 2)),  # Cluster 3
    np.random.normal(loc=[40, 35], scale=2.2, size=(20, 2)),  # Cluster 4
    np.random.normal(loc=[50, 45], scale=1.7, size=(20, 2))   # Cluster 5: high cost, high dist
])
data += np.random.normal(0, 0.5, data.shape)  # Add noise
print("Dataset shape:", data.shape)
print("Features: [[monthly_cost, distance]]")

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

sil_scores = []
inertias = []
best_k = 0
best_score = -1

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)  # Multiple inits
    labels = kmeans.fit_predict(data_scaled)
    sil = silhouette_score(data_scaled, labels)
    sil_scores.append(sil)
    inertias.append(kmeans.inertia_)
    print(f"For k={k}, Silhouette Score: {sil:.4f}")
    if sil > best_score:
        best_score = sil
        best_k = k

print(f"""
Best k: {best_k} with Silhouette Score: {best_score:.4f}""")

kmeans_final = KMeans(n_clusters=best_k, random_state=42, n_init=10)
labels_final = kmeans_final.fit_predict(data_scaled)

plt.figure(figsize=(10, 6))
scatter = plt.scatter(data[:, 0], data[:, 1], c=labels_final, cmap='viridis', s=50, alpha=0.7)
plt.xlabel('Monthly Transport Cost ($)')
plt.ylabel('Distance (km)')
plt.title(f"""KMeans Clustering (k={best_k}, Silhouette: {best_score:.4f})
Clusters shown by color""")
plt.colorbar(scatter, label='Cluster')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

centers_scaled = scaler.inverse_transform(kmeans_final.cluster_centers_)
print("""
Cluster Centers (Cost, Distance):""")
for i, center in enumerate(centers_scaled):
    print(f"Cluster {i}: {center}")
```
